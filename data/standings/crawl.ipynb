```python
import pandas as pd
import numpy as np
```

```python
club_df = pd.read_csv('../club/clubDB.csv')
club_df.head()
```

```python
standings = pd.DataFrame(columns=['lyear_id', 'club_id', 'match_play', 'points', 'pts', 'goal', 'goal_against'])
standings['club_id'] = club_df['club_id']
standings.head()
```

```python
for i in range(len(standings.index)):
    if standings['club_id'][i] <= 20 and standings['club_id'][i] >= 1:
        standings['lyear_id'][i] = 1 
    elif standings['club_id'][i] <=40 and standings['club_id'][i] >=21:
        standings['lyear_id'][i] = 125
    elif standings['club_id'][i] <=60 and standings['club_id'][i] >=41:
        standings['lyear_id'][i] = 160
    elif standings['club_id'][i] <= 80 and standings['club_id'][i] >= 61:
        standings['lyear_id'][i] = 195
    elif standings['club_id'][i]  <= 98 and standings['club_id'][i] >=81:
        standings['lyear_id'][i] = 223

standings.head()
```

```python
standings.fillna(0)
standings.head()
```

```python
standings.fillna(0).astype(int)
standings.head()
```

```python
standings[['match_play', 'points', 'pts', 'goal', 'goal_against']] = 0
standings.head()
```

```python
standings.to_csv('./standingsDB.csv', index=False, header=False)
```
